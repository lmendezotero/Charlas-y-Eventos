```python
# Importar librerias
import os
import pandas as pd
import numpy as np
from datetime import date
```

### 1. Tratamiento de datos de Ingresos

```python
df_HechosVentasDiario = pd.read_csv('HechosVentasDiario.csv')
df_HechosVentasDiario[['Grupo']] =  "7"
df_HechosVentasDiario[['Tipo']] =  "Ingreso"
df_HechosVentasDiario = df_HechosVentasDiario[['Fecha', 'Ejercicio', 'Mes',  'Grupo', 'Tipo', 'CuentaContable', 'DescripcionCuenta', 'ClienteID', 'ProductoID', 'Importe']]
df_HechosVentasDiario
```

```python
df_HechosOtrosIngresosNoOrdinariosDiario = pd.read_csv('HechosOtrosIngresosNoOrdinariosDiario.csv')
df_HechosOtrosIngresosNoOrdinariosDiario[['Grupo']] =  "7"
df_HechosOtrosIngresosNoOrdinariosDiario[['Tipo']] =  "Ingreso"
df_HechosOtrosIngresosNoOrdinariosDiario = df_HechosOtrosIngresosNoOrdinariosDiario[['Fecha', 'Ejercicio', 'Mes',  'Grupo', 'Tipo', 'CuentaContable', 'DescripcionCuenta', 'ClienteID', 'ProductoID', 'Importe']]
df_HechosOtrosIngresosNoOrdinariosDiario
```

```python
df_HechosOtrosIngresosExplotacionDiario = pd.read_csv('HechosOtrosIngresosExplotacionDiario.csv')
df_HechosOtrosIngresosExplotacionDiario[['Grupo']] =  "7"
df_HechosOtrosIngresosExplotacionDiario[['Tipo']] =  "Ingreso"
df_HechosOtrosIngresosExplotacionDiario = df_HechosOtrosIngresosExplotacionDiario[['Fecha', 'Ejercicio', 'Mes',  'Grupo', 'Tipo', 'CuentaContable', 'DescripcionCuenta', 'ClienteID', 'ProductoID', 'Importe']]
df_HechosOtrosIngresosExplotacionDiario
```

```python
df_fact_mov_ingresos =pd.concat([df_HechosVentasDiario,df_HechosOtrosIngresosNoOrdinariosDiario, df_HechosOtrosIngresosExplotacionDiario])
df_fact_mov_ingresos
```

```python
df_fact_mov_ingresos.to_csv("fact_mov_ingresos.csv", index=False)
```
