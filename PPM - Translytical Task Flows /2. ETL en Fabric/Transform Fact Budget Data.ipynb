```python
# Importar librerias
import os
import pandas as pd
import numpy as np
from datetime import date
```

### 1. Tratamiento de datos de Ingresos

```python
# Cargar datos de movimientos contables de ingresos
df_fact_mov_ingresos = pd.read_csv('fact_mov_ingresos.csv', sep=';')
# Visualizar resultados
df_fact_mov_ingresos
```

```python
# Generar columna de presupuesto
df_fact_mov_ingresos['Importe_Presupuesto'] = (
    df_fact_mov_ingresos['Importe'] * 
    (1 + np.random.uniform(-0.05, 0.05, size=len(df_fact_mov_ingresos)))
)
# Redondear decimales
df_fact_mov_ingresos['Importe_Presupuesto'] = df_fact_mov_ingresos['Importe_Presupuesto'].round(0).astype(int)

# Eliminar columna "Importe"
df_fact_mov_ingresos.drop(columns=['Importe'], inplace=True)

# Renombrar nombre columna Importe_Presupuesto por Importe
df_fact_mov_ingresos = df_fact_mov_ingresos.rename(columns={'Importe_Presupuesto': 'Importe'})

# Visualizar resultados
df_fact_mov_ingresos
```

```python
#df_fact_mov_ingresos.dtypes
```

```python
# Exportar resultados a csv
df_fact_mov_ingresos.to_csv("PresupuestoVentas2025.csv", index=False)
```

### 2. Tratamiento de datos de Gastos

```python
# Cargar datos de movimientos contables de gastos
df_fact_mov_gastos = pd.read_csv('fact_mov_gastos.csv', sep=';')
# Eliminar columnas innecesarias
df_fact_mov_gastos.drop(columns=['Unnamed: 0'], inplace=True)
# Visualizar resultados
df_fact_mov_gastos
```

```python
# Generar columna de presupuesto
df_fact_mov_gastos['Importe_Presupuesto'] = (
    df_fact_mov_gastos['Importe'] * 
    (1 + np.random.uniform(-0.05, 0.05, size=len(df_fact_mov_gastos)))
)
# Redondear decimales
df_fact_mov_gastos['Importe_Presupuesto'] = df_fact_mov_gastos['Importe_Presupuesto'].round(0).astype(int)

# Eliminar columna "Importe"
df_fact_mov_gastos.drop(columns=['Importe'], inplace=True)

# Renombrar nombre columna Importe_Presupuesto por Importe
df_fact_mov_gastos = df_fact_mov_gastos.rename(columns={'Importe_Presupuesto': 'Importe'})

# Visualizar resultados
df_fact_mov_gastos
```

```python
# Exportar resultados a csv
df_fact_mov_gastos.to_csv("PresupuestoGastos2025.csv", index=False)
```
