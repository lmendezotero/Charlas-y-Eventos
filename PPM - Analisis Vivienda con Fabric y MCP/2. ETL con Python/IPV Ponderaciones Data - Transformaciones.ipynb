```python
import pandas as pd
from pathlib import Path
```

```python
# Definición de rutas de ficheros
export_ine_path = Path(r"C:\Users\LorenaMéndez\Desktop\Analisis_Precios_vivienda\Export_INE")
ipv_25174_file_path = export_ine_path / "IPV - Export 25174 - Ponderaciones.xlsx"
```

```python
# Cargar fichero de exportación del indice del IPV
df_ipv_25174_gen = pd.read_excel(ipv_25174_file_path, sheet_name="tabla-25174-general")

# Usar la primera fila como nombres de columnas
df_ipv_25174_gen.columns = df_ipv_25174_gen.iloc[0]
# Eliminar la primera fila del DataFrame
df_ipv_25174_gen = df_ipv_25174_gen.iloc[1:]
# Resetear índice
df_ipv_25174_gen = df_ipv_25174_gen.reset_index(drop=True)

# Renombrar la primera columna del dataframe de pandas
df_ipv_25174_gen.rename(columns={df_ipv_25174_gen.columns[0]: "ref_geografica"}, inplace=True)

# Trasponer columnas de fechas por filas
df_ipv_25174_gen = df_ipv_25174_gen.melt(
    id_vars=["ref_geografica"],
    var_name="periodo",
    value_name="valor"
)

# Eliminar valores en blanco al inicio de columnas
df_ipv_25174_gen["ref_geografica"] = df_ipv_25174_gen["ref_geografica"].str.lstrip()

# Añadir columna de fecha
df_ipv_25174_gen["fecha"] = pd.to_datetime(
    dict(
        year=df_ipv_25174_gen["periodo"],
        month=1,
        day=1
    )
)

# Reordenar columnas
df_ipv_25174_gen = df_ipv_25174_gen[
    ["periodo", "fecha", "ref_geografica"] +
    df_ipv_25174_gen.columns.drop(["ref_geografica", "periodo", "fecha"]).tolist()
]

# Visualizar los resultados finales
df_ipv_25174_gen.head()
```

```python
# Exportar los resultados a excel
df_ipv_25174_gen.to_excel("ipv_25174_pond_gen.xlsx", index=False)
```

```python
# Mover el fichero al directorio de destino

# carpeta de origen
origen = Path("ipv_25174_pond_gen.xlsx")

# carpeta destino
output_dir = Path(
    r"C:\Users\LorenaMéndez\Desktop\Analisis_Precios_vivienda\Datos INE Procesados"
)
output_dir.mkdir(parents=True, exist_ok=True)

# ruta completa de destino
destino = output_dir / origen.name

# si existe en destino, lo eliminamos (para permitir sobrescritura)
if destino.exists():
    destino.unlink()

# mover el fichero (esto elimina el original de la raíz)
origen.rename(destino)
```
