## Cálculo Tabla Provincias INE

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Definición de rutas de ficheros del Registro de la Propiedad 
export_rp_path = Path(r"C:\Users\LorenaMéndez\Desktop\Analisis_Precios_vivienda\Data - Registro Propiedad")
comparativa_compraventa_file = export_rp_path / "Export-ComparativaVivienda-RP.xlsx"
```

```python
# Carga del fichero de comparativa de compraventa de vivienda
df_comparativa_compraventa_rp = pd.read_excel(comparativa_compraventa_file, sheet_name="Datos")

# Visualizar los resultados
df_comparativa_compraventa_rp
```

```python
#df_comparativa_compraventa_rp.columns
```

```python
## 1. Calculo tabla Geografia
df_geo_prev = df_comparativa_compraventa_rp[['geo']]
# Eliminar duplicados
df_geografia = df_geo_prev.drop_duplicates()
# Renombrar columnas
df_geografia = df_geografia.rename(columns={"geo": "geografia"})
#Añadir la columna id_geo al dataframe
map_geo = {
    "Nacional": 1,
    "Comunidad": 2,
    "Provincia": 3
}
df_geografia["id_geo"] = df_geografia["geografia"].map(map_geo).fillna(99).astype(int)
# Ordenar columnas
df_geografia = df_geografia[['id_geo','geografia']]
# Visualizar los resultados
df_geografia
```

```python
# Guardar resultado final de la tabla de geografia
df_geografia.to_excel("D_geografia_ine.xlsx", index=False)
df_geografia.to_csv("D_geografia_ine.csv", index=False)
```

```python
## 2. Calculo tabla provincias INE
df_provincias_ine_prev = df_comparativa_compraventa_rp[['cod-ca','cod-prv', 'prv']]
# Eliminar duplicados
df_provincias_ine_prev = df_provincias_ine_prev.drop_duplicates()
# Filtrar el DF para sacar filas de comunidad y nacional
df_provincias_ine = df_provincias_ine_prev[(df_provincias_ine_prev["cod-prv"] != 99) & (df_provincias_ine_prev["cod-prv"] != 98)]
# Renombrar columnas
df_provincias_ine = df_provincias_ine.rename(columns={"prv": "nombre_provincia"})
# Añdir columnas id_provincia y nombre_provincia_completo
df_provincias_ine["id_provincia"] = (
    df_provincias_ine["cod-ca"].astype(str).str.zfill(2)
    + "-"
    + df_provincias_ine["cod-prv"].astype(str).str.zfill(2)
)
df_provincias_ine["nombre_provincia_completo"] = (
    df_provincias_ine["id_provincia"].astype(str).str.zfill(2)
    + " - "
    + df_provincias_ine["nombre_provincia"]
)

# Ordenar el Dataframe
df_provincias_ine = df_provincias_ine[['id_provincia','cod-ca', 'cod-prv', 'nombre_provincia', 'nombre_provincia_completo']]
# Visualizar los resultados
df_provincias_ine
```

```python
#df_provincias_ine.columns
```

```python
# Guardar resultado final de la tabla de provincias
df_provincias_ine.to_excel("D_provincias_ine.xlsx", index=False)
df_provincias_ine.to_csv("D_provincias_ine.csv", index=False)
```
