## Tratamiento Datos Compraventa (RP)

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Definición de rutas de ficheros del Registro de la Propiedad 
export_rp_path = Path(r"C:\Users\LorenaMéndez\Desktop\Analisis_Precios_vivienda\Data - Registro Propiedad")
comparativa_compraventa_file = export_rp_path / "Export-ComparativaVivienda-RP.xlsx"
```

```python
# Carga del fichero de comparativa de compraventa de vivienda
df_comparativa_compraventa_rp_prev = pd.read_excel(comparativa_compraventa_file, sheet_name="Datos")

# Incluir la clave compuesta de id_provincia
df_comparativa_compraventa_rp_prev["id_provincia"] = (
    df_comparativa_compraventa_rp_prev["cod-ca"].astype(str).str.zfill(2)
    + "-"
    + df_comparativa_compraventa_rp_prev["cod-prv"].astype(str).str.zfill(2)
)

# Añadir la columna id_geo al dataframe
map_geo = {
    "Nacional": 1,
    "Comunidad": 2,
    "Provincia": 3
}
df_comparativa_compraventa_rp_prev["id_geo"] = df_comparativa_compraventa_rp_prev["geo"].map(map_geo).fillna(99).astype(int)

# Renombrar columnas
df_comparativa_compraventa_rp_prev = df_comparativa_compraventa_rp_prev.rename(columns={"ano": "ejercicio"})

# Cálculo de la columna de fecha
# Mapeo trimestre -> mes
map_trim_mes = {
    1: "01",
    2: "04",
    3: "07",
    4: "10"
}
# Crear columna fecha
df_comparativa_compraventa_rp_prev["fecha"] = pd.to_datetime(
    df_comparativa_compraventa_rp_prev["ejercicio"].astype(str)
    + "-"
    + df_comparativa_compraventa_rp_prev["trim"].map(map_trim_mes)
    + "-01"
)

# Visualizar los resultados
df_comparativa_compraventa_rp_prev
```

```python
# Filtrado de columnas finales en el DF
df_comparativa_compraventa_rp = df_comparativa_compraventa_rp_prev[['fecha','ejercicio', 'trim','id_geo','cod-ca','id_provincia',
                                    'viv-num','viv-imp', 'viv-pm2', 'gar-num', 'gar-imp', 'gar-pm2', 'tras-num',
                                       'tras-imp', 'tras-pm2']]
# Visualizar los resultados
df_comparativa_compraventa_rp
```
